```python
## plot ZH_HW
for i in np.arange(1, 3):
    plt.figure()
    plt.title("ZX_HW{}".format(i))
    for j in np.arange(1, 7):
        name = "ZX_HW{0}_{1}".format(i, j)
        plt.plot(data1[name], label=name)
    plt.legend(loc='best')
    plt.tight_layout()
```
